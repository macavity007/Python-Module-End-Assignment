<h3 style="color:green;"> 1. distribution of employees across each team and calculate the percentage split relative to the total 
number of employees</h3>

```python
disribution_emp=df["Team"].value_counts()
print("Distribution of Employees Each Team:\n",disribution_emp)
total_emp=len(df)
percentage_split=(disribution_emp/total_emp)*100
print("percentage of ech employee in team:\n",percentage_split)
```

```python
plt.figure(figsize=(9,8))
plt.pie( percentage_split,labels=percentage_split.index)
plt.title('Percentage Split of Employees Across Teams')
plt.show()

```

<h3 style="color:blue;">4.Discover which team and position have the highest salary expenditure</h3>

```python
s=df["Salary"].value_counts()
print(s)
l=s.idxmax()
l
```

in this code we find how many employee can get same salary ,947276 is salary 32 employee gets

```python
salary= df.groupby(['Team', 'Position'])['Salary'].sum()
highest=salary.idxmax()
print("Team and Position with the Highest Salary Expenditure:")
print("Team:", highest[0])
print("Position:", highest[1])
```

Group the data by team and position, then calculate total salary expenditure for each group, Find the team and position with the highest 
salary, function returns the index (which is a tuple containing the team and position) where the maximum value occurs in the salary.
finding the index of the maximum salary expenditure, we directly access the elements of the resulting tuple to retrieve the team and position
with the highest salary.highest[0] retrieves the team name, and highest[1] retrieves the position name from the tuple.
team  'LOS ANGELES LAKERS 'get highest salary expenditure and  postion 'SF' get highest salary expenditure 

<h4 style="color:blue;">graphical representation highest salary expenditure. </h4>

```python
team = highest[0]
position = highest[1]
expenditure = salary[highest]
plt.figure(figsize=(8, 4))
plt.barh(team, expenditure, color='lightblue')
plt.title('Highest Salary by Team')
plt.xlabel('Salary')
plt.ylabel('Team')
plt.show()

# position 
plt.figure(figsize=(8, 4))
plt.barh(position, expenditure, color='lightgreen')
plt.title('Highest Salary by Position')
plt.xlabel('Salary')
plt.ylabel('Position')
plt.show()
```

Assuming highest is a tuple containing the team and position with the highest salary expenditure
Assuming salary is a DataFrame or Series containing the highest salary expenditure by team and position, Plotting the horizontal bar chart for team

<h3 style="color:blue;"> 5.Investigate if there's any correlation between age and salary</h3>

```python
correlation = df['Age'].corr(df['Salary'])
print("Correlation between Age and Salary:", correlation)
```

calculate the correlation coefficient between these two variables. The correlation coefficient quantifies the strength and direction of 
the linear relationship between two variables. Here's how you can do it using Python with the corr() method.A correlation coefficient close to 
1 indicates a strong positive correlation (i.e., as one variable increases, the other variable tends to increase).
A correlation coefficient close to -1 indicates a strong negative correlation (i.e., as one variable increases, the other variable tends to decrease).
A correlation coefficient close to 0 indicates little to no linear relationship between the variables.

A value of 0.205 suggests a positive correlation between age and salary, meaning that as age increases, salary tends to increase, and vice versa.
However, the correlation is weak, indicating that the relationship between age and salary is not very strong. 

<h4 style="color:green;">Graphical Representation</h4>
